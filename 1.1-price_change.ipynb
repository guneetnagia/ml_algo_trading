```python
# Visualize the data
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
close = pd.read_csv("close.csv", index_col = "Date", parse_dates= ["Date"])
close
```

```python
msft = close.MSFT.dropna().to_frame().copy()
```

```python
msft
```

```python
msft.rename(columns={"MSFT":"Price"}, inplace=True)
```

```python
msft
```

```python
msft["P_lag1"] = msft.shift(periods = 1)
msft
# shift price by 1 day, so that it is wasy to compare
```

```python
# Absolute Price Changes (Difference)
msft["P_diff"] = msft.Price.sub(msft.P_lag1) # Alternate 1
```

```python
msft
```

```python
msft["P_diff2"] = msft.Price.diff(periods=1) # Alternate 2
msft
```

```python
## Relative / Percentage Price changes (returns)
```

```python
msft.Price.div(msft.P_lag1) - 1 #Alternate 1
```

```python
msft["Returns"] = msft.Price.pct_change(periods=1) # Alternate 2
msft
```

```python
msft.drop(columns=["P_lag1","P_diff","P_diff2"], inplace = True)
```
